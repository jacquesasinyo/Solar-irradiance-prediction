```python
# Select a specific time range for training and testing
train_start_year, train_end_year = 2016, 2017  # Consider only the first two years
test_year = 2018  # The following year

# Filter data based on the specified time range
train_data = df[(df['Year'] >= train_start_year) & (df['Year'] <= train_end_year)]
test_data = df[df['Year'] == test_year]

# Randomly sample tuples for training and testing
train_size = 100000
test_size = 20000

# Set a random seed for reproducibility
np.random.seed(42)

# Randomly sample training data
selected_indices_train = np.random.choice(train_data.index, train_size, replace=False)
train_sample = train_data.loc[selected_indices_train]

# Randomly sample testing data
selected_indices_test = np.random.choice(test_data.index, test_size, replace=False)
test_sample = test_data.loc[selected_indices_test]
```

```python
# List of features with outliers to be winsorized
features_with_outliers = ['Avg Wind Speed @ 42ft [m/s]', 'Peak Wind Speed @ 42ft [m/s]']

features_to_normalize = ['Air Temperature [deg C]', 'Rel Humidity [%]', 'Est Pressure [mBar]',
                          'Precipitation [mm]', 'Precipitation (Accumulated) [mm]']

# Apply winsorization to selected features to handle outliers by clipping extreme values. 
for feature in features_with_outliers:
    train_sample[feature] = winsorize(train_sample[feature], limits=[0.05, 0.05])
    test_sample[feature] = winsorize(test_sample[feature], limits=[0.05, 0.05])


# Apply normalization to scale all features to a similar range (between 0 and 1.) 
scaler = MinMaxScaler()
X_train[features_to_normalize] = scaler.fit_transform(X_train[features_to_normalize])

X_test[features_to_normalize] = scaler.transform(X_test[features_to_normalize])
```

```python
# Define the parameter grid for SVM
param_grid_svm = {
    'C': [10],
    'kernel': ['rbf'],
    'gamma': ['auto']
}

# Create the SVR model
svm_model = SVR()

# Create GridSearchCV
grid_search_svm = GridSearchCV(estimator=svm_model, param_grid=param_grid_svm, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

# Fit the grid search to the data
grid_search_svm.fit(X_train_transformed, y_train)

# Get the best parameters
best_params_svm = grid_search_svm.best_params_
print("Best Hyperparameters for SVM:", best_params_svm)

# Use the best model for predictions
best_svm_model = grid_search_svm.best_estimator_
svm_predictions = best_svm_model.predict(X_test_transformed)
```

```python
# Define the LSTM model as a function
def create_lstm_model(units=5, activation='relu', learning_rate=0.001):
    model = Sequential()
    model.add(LSTM(units, activation=activation, input_shape=(1, X_train_transformed.shape[1])))
    model.add(Dense(1))
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate), loss='mse')
    return model

# Define the parameter grid for LSTM
param_grid_lstm = {
    'units': [5, 10, 15],              # Number of LSTM units
    'activation': ['relu', 'elu'],     # Activation function
    'learning_rate': [0.001, 0.01, 0.1] # Learning rate
}

# Create the LSTM model
lstm_model = KerasRegressor(build_fn=create_lstm_model, epochs=50, batch_size=32, verbose=0)

# Create GridSearchCV
grid_search_lstm = GridSearchCV(estimator=lstm_model, param_grid=param_grid_lstm, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

# Reshape the data for LSTM input
X_train_reshaped = X_train_transformed.reshape((X_train_transformed.shape[0], 1, X_train_transformed.shape[1]))

# Fit the grid search to the data
grid_search_lstm.fit(X_train_reshaped, y_train)

# Get the best parameters
best_params_lstm = grid_search_lstm.best_params_
print("Best Hyperparameters for LSTM:", best_params_lstm)

# Use the best model for predictions
best_lstm_model = grid_search_lstm.best_estimator_
X_test_reshaped = X_test_transformed.reshape((X_test_transformed.shape[0], 1, X_test_transformed.shape[1]))
lstm_predictions = best_lstm_model.predict(X_test_reshaped)
```
